```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

### basic graph

```python
x = [0,1,2,3]
y = [1,2,3,4]
plt.plot(x, y, label='2x',markeredgecolor='red', linewidth=2, marker='.', markersize=10,linestyle='--')
plt.xticks([1,2,3])
plt.yticks([1,2,4])
plt.legend()

plt.show()
```

```python
x = [1,2]
y = [3,4]
X, Y = np.meshgrid(x,y)
Z = X*Y
print(X)
print(Y)
print(Z)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
```
