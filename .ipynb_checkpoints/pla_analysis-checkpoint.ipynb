```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)

def display_data(X, labels):
    X0 = X[:, labels[0]==-1]
    X1 = X[:, labels[0]==1]
    plt.plot(X0[0,:], X0[1,:], 'b^')
    plt.plot(X1[0,:], X1[1,:], 'ro')
#plt.show()
    
means = [[1,2], [4,5]]
cov = [[1,0], [0,.2]]
N = 100
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
original_labels = np.asarray([-1]*N + [1]*N)
X = np.concatenate((X0, X1), axis = 0)
#display_data(X, original_labels)
X = np.concatenate((X0,X1),axis=0).T
y = np.concatenate((np.ones((N,1)), -1*np.ones((N,1))), axis=0).T
ones = np.ones((1, X.shape[1]))
Xbar = np.concatenate((ones, X), axis = 0)
```

```python
def h(w,x):
    return np.sign(np.dot(w.T, x))

def has_converged(X, w, y):
    return np.array_equal(h(w,X), y)

def pla(X, y, w_init):
    w = [w_init]
    d = X.shape[0]
    N = X.shape[1]
    eta = 1 
    max_cnt = 20000
    cnt = 0
    
    """
    có dạng linear thì dùng dạng input là ma trận với mỗi vector cột là 1 điểm dữ liệu
    
    B1: init w_init
    B2: áp dụng SGD
    B3: kiểm tra nếu bị phân lớp sai thì cập nhật w = w + xi*yi
    B4: nếu hội tụ thì dừng or lặp lại bước 2
    """
    
    while cnt < max_cnt:
        mix_id = np.random.permutation(N)
        for i in mix_id:
            xi = X[:, mix_id[i]].reshape(-1,1)
            yi = y[0, mix_id[i]] # scalar
            if h(w[-1], xi)[0] != yi:
                w_new = w[-1] + eta*xi*yi
                w.append(w_new)
        if has_converged(X, w[-1], y):
            break
    return w

w_init = np.random.randn(Xbar.shape[0], 1)
#print(w_init)
#print(Xbar.shape)
w = pla(Xbar, y, w_init)
display_data(X, y)

w_ = w[-1]
w0 = w_[0][0]
w1 = w_[1][0]
w2 = w_[2][0]

# w1*x + w2*y + w0 = 0
xx = np.linspace(-2, 8, 2)
yy = (-w0 - w1*xx)/w2

plt.plot(xx, yy, 'g')

plt.show()
```
