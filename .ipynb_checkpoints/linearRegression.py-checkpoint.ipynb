```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T

plt.plot(X, y, 'ro')
plt.axis('equal')
plt.show()
```

```python
ones = np.ones((1,len(X))).T
Xbar = np.concatenate((ones, X), axis=1)
#print(Xbar)
```

```python
A = Xbar.T.dot(Xbar)
b = Xbar.T.dot(y)
w = np.linalg.pinv(A).dot(b)
print(w)
```

```python
w0 = w[0][0]
w1 = w[1][0]
x = np.linspace(145, 190, 2)
y = w0 + w1*x
plt.plot(x, y, 'g', color='blue')

X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
labels = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T

plt.plot(X, labels, 'ro')
plt.show()
```

```python
from sklearn import linear_model
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
labels = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
regr = linear_model.LinearRegression(fit_intercept=False)
regr.fit(Xbar, labels)
print(regr.coef_)
```

```python
print(w.T)
print(regr.coef_)
print(np.linalg.norm(w.T - regr.coef_))
```
