```python
import numpy as np
import matplotlib.pyplot as plt

def display_data(X, labels):
    X0 = X[labels==0, :]
    X1 = X[labels==1, :]
    plt.plot(X0[:,0], X0[:,1], 'go')
    plt.plot(X1[:,0], X1[:,1], 'b^')
    plt.show()

np.random.seed(2)
means = [[1,2], [4,7]]
cov = [[1,0], [0,1]]

N = 50
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)

X = np.concatenate((X0,X1), axis=0)
#ones = np.ones((1, X.shape[1]))
#Xbar = np.concatenate((ones, X), axis=0)
o_labels = np.asarray([0]*N + [1]*N)
#print(X.shape)

display_data(X, o_labels)
```

```python
from scipy.spatial.distance import cdist
def init_center(X, cluster=2):
    # one row match with one point
    return X[np.random.choice(X.shape[0], cluster, replace=False)]

def assign_label(X, center):
    d = cdist(X, center)
    return np.argmin(d, axis=1)

def update_centers(X, labels, cluster):
    new_center = np.zeros((cluster, X.shape[1]))
    for j in range(cluster):
        xj = X[labels==j, :]
        new_center[j] = np.mean(xj, axis=0)
    return new_center

def has_converged(centers, old_centers):
    return set([tuple(a) for a in centers]) == set([tuple(a) for a in old_centers])

def kmeans(X, cluster):
    cluster = 2 
    centers = [init_center(X, cluster)] # list of center point
    labels = []
    #print(X)
    #print(centers[-1])
    
    while True:
        labels.append(assign_label(X, centers[-1]))
        new_center = update_centers(X, labels[-1], cluster)
        if has_converged(new_center, centers[-1]):
            break
        centers.append(new_center)
    
    return (centers, labels)
       
""" 
option 2: 
def has_converge(labels, old_labels)
"""
centers, labels = kmeans(X, 2)
```

```python
display_data(X, labels[-1])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
lb = kmeans.predict(X)
display_data(X, lb)
```
