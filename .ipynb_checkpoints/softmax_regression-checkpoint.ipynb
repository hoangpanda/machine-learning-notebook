```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse
```

```python
np.random.seed(1)
```

```python
means = [[2, 2], [8, 3], [3, 6]]
cov = [[1, 0], [0, 1]]
```

```python
N = 500
```

```python
np.random.seed(1)
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
X2 = np.random.multivariate_normal(means[2], cov, N)
X = np.concatenate((X0,X1,X2), axis = 0)
X = np.concatenate((np.ones((3*N,1)),X), axis=1).T
original_labels = np.asarray([0]*N + [1]*N + [2]*N)
print(X)
```

```python
def data_display(X, labels):
    X0 = X[:, labels==0]
    X1 = X[:, labels==1]
    X2 = X[:, labels==2]
    plt.plot(X0[0, :], X0[1, :], 'rs', alpha=.4, markersize=4)
    plt.plot(X1[0, :], X1[1, :], 'b^', alpha=.4, markersize=4)
    plt.plot(X2[0, :], X2[1, :], 'go', alpha=.4, markersize=4)
    plt.show()
```

```python
data_display(X[1:, :], original_labels)
```

```python
def convert_label(y, c):
    # data row column
    return sparse.coo_matrix((np.ones_like(y), (y, np.arange(len(y)))), shape=(c, len(y))).toarray()
```

```python
c = 3
y = convert_label(original_labels, c)
print(y)
```
